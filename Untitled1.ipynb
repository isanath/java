Adjusting the photo urls and finding their count

```python
for i in range(len(df3['photo_urls'])):
    if (isinstance(df3['photo_urls'][i],float) or (len(df3['photo_urls'][i])<6)):
        df3['photo_urls'][i]=0
    else:
        df3['photo_urls'][i]=df3['photo_urls'][i].replace('title','"title',1)
        df3['photo_urls'][i]=df3['photo_urls'][i][:-1]
        df3['photo_urls'][i]=df3['photo_urls'][i].replace("\\","")
        df3['photo_urls'][i]=len(json.loads(df3['photo_urls'][i]))
        
```
